```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import MarkerCluster

# Example dataset
data = pd.DataFrame({
    'Case ID': [1, 2, 3, 4, 5],
    'Victim(s)': ['David Faraday, Betty Lou Jensen', 'Darlene Ferrin, Michael Mageau', 'Bryan Hartnell, Cecelia Ann Shepard', 'Cheri Jo Bates', 'Paul Stine'],
    'Date': ['12/20/1968', '7/4/1969', '9/27/1969', '10/30/1966', '10/11/1969'],
    'Weapon Used': ['.22 caliber gun', '9mm semi-automatic gun', 'Knife', 'Knife', '.38 caliber gun'],
    'Status': ['Deceased', '1 Dead, 1 Survived', '2 Deceased', 'Deceased', 'Deceased'],
    'Location': ['Lake Herman Road, California', 'Blue Rock Springs Park, CA', 'Lake Berryessa, Napa County, CA', 'Riverside, California', 'Presidio Heights, San Francisco, CA'],
    'Geographical Coordinates': ['38.1232° N, 122.2447° W', '38.1506° N, 122.2353° W', '38.2943° N, 122.2276° W', '33.9736° N, 117.3693° W', '37.7987° N, 122.4480° W'],
    'Victim Profile': ['Ages 17 (male) and 16 (female), students', 'Ages 22 (female) and 19 (male), friends', 'Ages 20 (male) and 22 (female), friends', 'College student, age 18', 'Adult male, age 29'],
    'Modus Operandi (MO)': ['Targeted secluded lovers\' lanes, nighttime attacks', 'Repeated calls to police, taunting letters following crimes', 'Public space, random attack', 'No clear motive', 'Random attack, targeted a lone individual']
})

# Convert 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'])
data['Year'] = data['Date'].dt.year
```

```python
1. Crime Timeline Analysis (Attacks per Year)
plt.figure(figsize=(8, 6))
crime_counts = data.groupby('Year').size()
crime_counts.plot(kind='bar', color='#e74c3c')  # Red theme for the Zodiac Killer
plt.title('Zodiac Killer Attacks Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.tight_layout()
plt.show()
```

```python
# 1. Crime Timeline: Attacks per year
plt.figure(figsize=(8, 6))
crime_counts = data.groupby('Year').size()
crime_counts.plot(kind='bar', color='darkred')
plt.title('Zodiac Killer Attacks Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.tight_layout()
plt.show()



```

```python
# 2. Geographical Distribution of Crimes (with Details)
# Create map with enhanced details
map_center = [37.7749, -122.4194]  # San Francisco as central reference
crime_map = folium.Map(location=map_center, zoom_start=7)

# Add MarkerCluster for better interaction
marker_cluster = MarkerCluster().add_to(crime_map)

locations = [
    {'name': 'Lake Herman Road', 'coords': [38.1232, -122.2447], 'details': 'First confirmed Zodiac attack, two victims shot in a car.'},
    {'name': 'Blue Rock Springs Park', 'coords': [38.1506, -122.2353], 'details': 'Second attack, survivor provided suspect description.'},
    {'name': 'Lake Berryessa', 'coords': [38.2943, -122.2276], 'details': 'Victims stabbed while relaxing, Zodiac wore hood and glasses.'},
    {'name': 'Riverside', 'coords': [33.9736, -117.3693], 'details': 'Cheri Jo Bates found dead; Zodiac letters linked but unproven.'},
    {'name': 'Presidio Heights', 'coords': [37.7987, -122.4480], 'details': 'Paul Stine, a cab driver, shot and Zodiac interacted with police nearby.'}
]

# Adding markers with additional details
for loc in locations:
    folium.Marker(location=loc['coords'], popup=f"{loc['name']}<br>{loc['details']}", icon=folium.Icon(color='darkred')).add_to(marker_cluster)

# Save the detailed map
crime_map.save('zodiac_detailed_crimes_map.html')
```

```python
# 3. Weapon Distribution
plt.figure(figsize=(8, 6))
weapon_data = ['.22 caliber gun', '9mm semi-automatic gun', 'Knife', 'Knife', '.38 caliber gun']
weapon_counts = pd.Series(weapon_data).value_counts()
weapon_counts.plot(kind='pie', autopct='%1.1f%%', colors=['red', 'blue', 'green', 'yellow', 'purple'])
plt.title('Weapon Distribution in Zodiac Killer Crimes')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# 4. Victim Age Distribution Analysis
plt.figure(figsize=(8, 6))
victim_ages = [17, 16, 22, 19, 20, 22, 18, 29]
sns.histplot(victim_ages, kde=True, color='#8e44ad')  # Purple to match the theme
plt.title('Age Distribution of Zodiac Killer Victims')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
 #5. Modus Operandi (MO) Analysis (Target Types)
plt.figure(figsize=(8, 6))
mo_data = ['Couple', 'Couple', 'Couple', 'Individual', 'Individual']
mo_counts = pd.Series(mo_data).value_counts()
mo_counts.plot(kind='bar', color='#2ecc71')  # Green for the nature of the crime
plt.title('Zodiac Killer Modus Operandi')
plt.xlabel('Target Type')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#6. Investigative Outcome Analysis
plt.figure(figsize=(8, 6))
status_data = ['Deceased', '1 Dead, 1 Survived', '2 Deceased', 'Deceased', 'Deceased']
status_counts = pd.Series(status_data).value_counts()
status_counts.plot(kind='bar', color='#f39c12')  # Orange color for investigation results
plt.title('Zodiac Killer Investigation Outcomes')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#Create a pie chart showing gender distribution of the victims
gender_data = ['Male', 'Female', 'Male', 'Female', 'Male']
gender_counts = pd.Series(gender_data).value_counts()
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['#3498db', '#f39c12'])
plt.title('Gender Distribution of Zodiac Killer Victims')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
